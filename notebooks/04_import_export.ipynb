# Core Data Structures

Pandas has two primary data structures:

- Series: Think of this as a single column in Excel. It is 1-dimensional (just a list of items), but unlike a standard Python list, every item has a specific label (called an Index).

- DataFrame: Think of this as the entire spreadsheet. It is 2-dimensional (rows and columns). A DataFrame is essentially just multiple Series joined together.

```python
# Series Example
import pandas as pd

s = pd.Series([10, 20, 30, 40])
s
```

```python
# DataFrame Example
data = {
    "customer": ["Anna", "David", "Liza"],
    "revenue": [100, 250, 180]
}

df = pd.DataFrame(data)
df
```
