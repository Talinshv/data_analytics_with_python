```python
sales = [120, 250, 80, 310, 95]
```

```python
# Compute total revenue manually using a loop.
total_revenue = 0
for value in sales:
    total_revenue = total_revenue + value
total_revenue
```

```python
# Compute average revenue.
average_revenue = total_revenue / len(sales)
average_revenue
```

```python
# Identify the maximum value without using max().
max_value = 0
for value in sales:
    if value > max_value:
        max_value = value
max_value
```

```python
# Count how many values are greater than 150.
grater_than_150 = 0
for value in sales:
    if value >150:
        grater_than_150 = grater_than_150 + 1
grater_than_150
```

```python
# Create a new list of revenues after adding 5% commission.
revenues_with_commission = [r *1.05 for r in sales]
revenues_with_commission
```

# Part 3

Segmentation rules:

- “Premium” if revenue > 250
- “Standard” if 100 < revenue ≤ 250
- “Low” otherwise

```python
# Create a list of segment labels.
segments = ["premium" if r > 250 else "standard" if r > 100 and r <= 250 else "low" for r in revenues]
segments
```

```python
# Count how many customers fall into each segment.
count_customers = [0, 0, 0]
for segment in segments:
    if segment == "premium":
        count_customers[0] = count_customers[0] + 1
    elif segment == "standard":
        count_customers[1] = count_customers[1] + 1
    else:
        count_customers[2] = count_customers[2] + 1
count_customers 
```

Add Markdown explanation: Why does the order of elif statements matter?

Python evaluates conditional statements from top to bottom. Once it finds a condition that is True, it executes that block and skips the rest of the chain.

If we had put the "Standard" check (r > 100) before the "Premium" check (r > 250), a revenue of 310 would trigger the r > 100 condition and be labeled "Standard." Python would never even look at the "Premium" rule. To get accurate results, you must generally order your conditions from the most specific (or highest value) to the most general (or lowest value).

Explain in Markdown why sets are unordered.


Sets are unordered because they are implemented using a data structure called a Hash Table.Unlike a List, which stores items in a specific sequence (like passengers in a row of seats), a Set uses a "hashing function" to assign each element a unique "address" in memory. This allows the computer to find an item almost instantly because it knows the exact address, rather than searching through the whole list. Because the goal is speed and uniqueness rather than sequence, the order in which items appear is not preserved.

Add Markdown explanation: Why are dictionaries useful for structured data?

Dictionaries are essential for structured data because they use key-value pairs, which allow us to label our data. Unlike a list where you have to remember that "index 0 is the name" and "index 1 is the revenue," a dictionary explicitly links a label (the key) to its data (the value).

This makes the data:

Self-describing: It's immediately clear what each piece of information represents.

Efficient: You can look up a specific value instantly using its key without searching through a whole list.

Flexible: You can easily add or remove fields (keys) for one record without breaking the structure of others.
