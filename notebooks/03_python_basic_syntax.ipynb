## Tuple ( )

Tuple is a specific way of grouping data together in Python. 
While it looks similar to a standard list, it has one superpower: immutability.
A tuple is a collection of items that are ordered and unchangeable. 
In code, you recognize them because they use parentheses () instead of square brackets [].
coordinates = (40.18, 44.51)  This stores two floating-point numbers together as a single unit (likely Latitude and Longitude).
* Core Properties
  - Ordered: The items stay in the exact sequence you put them in.
  - Indexed: You can grab specific items by their position (e.g., "Give me the first item in the tuple").
  - Immutable: This is the big one. Once a tuple is created, you cannot change, add, or remove items. It is "carved in stone."

```python
coordinates = (40.18, 44.51)
coordinates
```

## Set { }

Set is a collection data structure that behaves very differently from lists or tuples. 
Its main purpose is to handle unique data and perform mathematical logic.
* Unlike a list, a set doesn't care about the sequence. 
  If you print a set, the items might come out in a different order than you typed them.
* Because there is no fixed order, you cannot ask for "the first item" in a set.

```python
customer_ids = {1, 2, 3, 3, 4}
customer_ids
```

### Key Properties (The "Rules")

* Unordered: There is no "first" or "last" item. Because of this, you can't use an index (like s1[0]) to grab a value.
* Unique values only: If you try to add 2 to a set that already has a 2, nothing happens. It's self-cleaning.
* Mutable container: Even though the items inside must be unchangeable (hashable), the set itself can grow or shrink as you add or remove elements.
* Fast membership checking: This is the set's "killer feature." Because it uses hashing, checking if an item exists (x in set) is nearly instantaneous, regardless of whether the set has 10 items or 10 million.

### Creating Sets

```python
s1 = {1, 2, 3}   # creating
s1
```

```python
s1 = {1, 2, 2, 3}   # creating + cleaning
s1
```

You are telling Python directly to create a set containing those specific numbers. It is slightly faster because Python sees the curly braces and knows exactly what to do immediately.

```python
s2 = set([1, 2, 2, 3])  # create a set from a list
s2
```

You are calling the set() function. To do this, you first created a list ([1, 2, 2, 3]) and then asked Python to convert that list into a set. This is a two-step process.

* Use { } when you already know the data you want to put in the set. It’s cleaner and more "Pythonic."
* Use set() when you have an existing collection (like a list or a tuple) that you want to transform into a set—usually to remove duplicates or to make membership testing faster.

### Set Algebra

It allows you to compare two different groups of data (segments) and find out how they overlap or differ.

#### Intersection (&)

```python
segment_a = {1, 2, 3}
segment_b = {3, 4, 5}
```

```python
# Finds only the elements that appear in both sets.
segment_a.intersection(segment_b)
# or
segment_a & segment_b
```

#### Union (|)

```python
# Combines both sets into one, keeping only unique elements.
segment_a.union(segment_b)
# or
segment_a | segment_b
```

#### Difference (-)

```python
# elements in A but not in B
segment_a.difference(segment_b)
# or
segment_a - segment_b
```

#### Symmetric Difference (^)

```python
# It keeps everything except the ones they have in common.
segment_a.symmetric_difference(segment_b)
# or
segment_a ^ segment_b
```

#### Analytical Use Cases

##### Removing Duplicates


```python
# Converting the list to a set automatically "de-duplicates" the data.
transactions = [101, 102, 102, 103, 104]
unique_transactions = set(transactions)
unique_transactions
```

##### Customer Overlap Between Campaigns (Intersection)

```python
campaign_a = {1, 2, 3, 4}
campaign_b = {3, 4, 5, 6}

overlap = campaign_a & campaign_b
overlap
```

##### Finding New Customers (Difference)

```python
new_customers = campaign_b - campaign_a
new_customers
```

##### Data Validation / Integrity Checks

```python
valid_ids = {1, 2, 3, 4, 5}
incoming_ids = {2, 3, 6}

invalid_ids = incoming_ids - valid_ids
invalid_ids
```

6 is flagged as invalid because it doesn't exist in the master list.

#### Limitations

##### No Indexing or Guaranteed Order

In a list, items are like people standing in a line (everyone has a number). 
In a set, items are like balls in a bag; you can check if a ball is in there, but there is no "first" ball.
In a set You cannot write s[0] to get the first item.
