## Tuple ( )

Tuple is a specific way of grouping data together in Python.

While it looks similar to a standard list, it has one superpower: immutability.

A tuple is a collection of items that are ordered and unchangeable. 

In code, you recognize them because they use parentheses () instead of square brackets [].

coordinates = (40.18, 44.51)  This stores two floating-point numbers together as a single unit (likely Latitude and Longitude).

* Core Properties
  - Ordered: The items stay in the exact sequence you put them in.
  - Indexed: You can grab specific items by their position (e.g., "Give me the first item in the tuple").
  - Immutable: This is the big one. Once a tuple is created, you cannot change, add, or remove items. It is "carved in stone."

```python
coordinates = (40.18, 44.51)
coordinates
```

## Set { }

Set is a collection data structure that behaves very differently from lists or tuples.
 
Its main purpose is to handle unique data and perform mathematical logic.
* Unlike a list, a set doesn't care about the sequence. 
  If you print a set, the items might come out in a different order than you typed them.
* Because there is no fixed order, you cannot ask for "the first item" in a set.

```python
customer_ids = {1, 2, 3, 3, 4}
customer_ids
```

### Key Properties (The "Rules")

* Unordered: There is no "first" or "last" item. Because of this, you can't use an index (like s1[0]) to grab a value.
* Unique values only: If you try to add 2 to a set that already has a 2, nothing happens. It's self-cleaning.
* Mutable container: Even though the items inside must be unchangeable (hashable), the set itself can grow or shrink as you add or remove elements.
* Fast membership checking: This is the set's "killer feature." Because it uses hashing, checking if an item exists (x in set) is nearly instantaneous, regardless of whether the set has 10 items or 10 million.

### Creating Sets

```python
s1 = {1, 2, 3}   # creating
s1
```

```python
s1 = {1, 2, 2, 3}   # creating + cleaning
s1
```

You are telling Python directly to create a set containing those specific numbers.
 
It is slightly faster because Python sees the curly braces and knows exactly what to do immediately.

```python
s2 = set([1, 2, 2, 3])  # create a set from a list by cleaning it
s2
```

You are calling the set() function. To do this, you first created a list ([1, 2, 2, 3]) and then asked Python to convert that list into a set. This is a two-step process.

* Use { } when you already know the data you want to put in the set. It’s cleaner and more "Pythonic."
* Use set() when you have an existing collection (like a list or a tuple) that you want to transform into a set—usually to remove duplicates or to make membership testing faster.

### Basic Operations

#### 1- Handling Elements (The "Editing" Tools)

Since sets are mutable, you can modify them after they are created using these specific commands:

```python
s = {1, 2, 3}
```

```python
s.add(4)        # Add element
s
```

```python
s.remove(2)     # Remove element (error if not found)
s
```

```python
# This is the "safe" version of remove. 
# It tries to delete 10, but if 10 isn't there, it just stays quiet and does nothing instead of crashing.
s.discard(10)   # Safe remove (no error)
s
```

```python
s.pop()         # Remove random element
s
```

```python
s.pop()         # Remove random element
s
```

#### 2-Membership Check (The "Question" Tools)

This is where sets shine. Using the in keyword, you can instantly check if a piece of data exists within the collection.

```python
3 in s   # True/false test
```

```python
4 in s # True/false test
```

### Set Algebra

It allows you to compare two different groups of data (segments) and find out how they overlap or differ.

#### Intersection (&)

```python
segment_a = {1, 2, 3}
segment_b = {3, 4, 5}
```

```python
# Finds only the elements that appear in both sets.
segment_a.intersection(segment_b)
# or
segment_a & segment_b
```

#### Union (|)

```python
# Combines both sets into one, keeping only unique elements.
segment_a.union(segment_b)
# or
segment_a | segment_b
```

#### Difference (-)

```python
# elements in A but not in B
segment_a.difference(segment_b)
# or
segment_a - segment_b
```

#### Symmetric Difference (^)

```python
# It keeps everything except the ones they have in common.
segment_a.symmetric_difference(segment_b)
# or
segment_a ^ segment_b
```

#### Analytical Use Cases

##### Removing Duplicates


```python
# Converting the list to a set automatically "de-duplicates" the data.
transactions = [101, 102, 102, 103, 104]
unique_transactions = set(transactions)
unique_transactions
```

##### Customer Overlap Between Campaigns (Intersection)

```python
campaign_a = {1, 2, 3, 4}
campaign_b = {3, 4, 5, 6}

overlap = campaign_a & campaign_b
overlap
```

##### Finding New Customers (Difference)

```python
new_customers = campaign_b - campaign_a
new_customers
```

##### Data Validation / Integrity Checks

```python
valid_ids = {1, 2, 3, 4, 5}
incoming_ids = {2, 3, 6}

invalid_ids = incoming_ids - valid_ids
invalid_ids
```

6 is flagged as invalid because it doesn't exist in the master list.

#### Limitations

##### No Indexing or Guaranteed Order

In a list, items are like people standing in a line (everyone has a number). 
In a set, items are like balls in a bag; you can check if a ball is in there, but there is no "first" ball.
In a set You cannot write s[0] to get the first item.

##### The "Hashable" Rule

Elements inside a set must be immutable (unchangeable).
To keep membership checking lightning-fast, Python creates a unique "fingerprint" (a hash) for every item. 
If an item can change (like a list can), its fingerprint would change too, and the set would "lose" it.

This will fail
s = {[1, 2], [3, 4]}

```python
# correct type
s = {(1, 2), (3, 4)}
```

#### When to Use Sets

* Order doesn't matter: You don't need a "first, second, third" sequence.

* Uniqueness is mandatory: You want to automatically ban duplicates.

* Speed is a priority: You need to check for an item's existence instantly.

* Group Logic: You need to compare two groups (like finding people who are in Group A but not Group B).

##### Sets as a "Mathematical Tool"

shifts the perspective from sets being just "boxes for data" to being "engines for logic." 
Because they are built on mathematical set theory, they enable:

* Efficient Deduplication: Cleaning messy data with one command.
* Group Comparisons: Using Intersection, Union, and Difference to segment audiences.

##### Analytical Real-World Use

where you'll actually see this in a data career:
* Customer Segments: "Show me users who bought a subscription AND a gift card."
* Campaign Overlaps: Measuring how much your Facebook ads and Google ads hit the same people.
* Data Validation: Checking if a list of new IDs contains any "bad" IDs that shouldn't be there.

Think of sets as your go-to structure for “who is in / who is not in” problems.

## Dictionary {}

Dictionary (or dict), is one of the most powerful and frequently used data structures in Python. 
If a list is a numbered sequence of items, a dictionary is a labeled collection of information.
It works just like a real-world dictionary: you look up a Key (the word) to find its Value (the definition).

### Structure: Keys and Values

A dictionary is made of Key-Value pairs.

customer = {"name": "Anna", "revenue": 150, "city": "Yerevan"}
 * "name", "revenue", and "city" are the Keys.
 * "Anna", 150, and "Yerevan" are the Values.

```python
customer = {
    "name": "Anna",
    "revenue": 150,
    "city": "Yerevan"
}

customer
```

### Core Properties

* Unique Keys: You can't have two keys with the same name (e.g., you can't have two "name" labels for one customer).

* Flexible Values: As you can see, the values can be different types: a string ("Anna"), an integer (150), or even another dictionary!

* Fast Lookup: Like a set, dictionaries are optimized so that finding a value by its key is nearly instantaneous, even with millions of entries.

### How to Use a Dictionary

```python
# Accessing Data
#Instead of using a number index like [0], you use the key name in square brackets:

customer["revenue"]
```

```python
# Adding data 
customer["segment"] = "Premium"
# Updating data
customer["revenue"] = 200
customer
```

```python
# Removing Data
del customer["city"]
customer
```

## From Dictionary to Structured Data

```python
customers = [
    {"name": "Anna", "revenue": 150},
    {"name": "David", "revenue": 220},
    {"name": "Mariam", "revenue": 90}
]

customers
```

### The Anatomy of Structured Data
Look closely at how the customers variable is built:

* The Outer Layer (The List): Notice the square brackets [ ]. This is a List used to hold a sequence of multiple items.

* The Inner Layer (The Dictionaries): Inside that list, each item is a Dictionary { }. Each dictionary represents one specific row or record (e.g., one person).

### Converting Code to a Table
* The Keys in the dictionaries ("name", "revenue") become the Column Headers.
* The Values ("Anna", 150) become the Cells.
* Each Dictionary in the list becomes a Row.

### Why This Matters: The Path to Pandas
Pandas is the most popular Python library for data analysis. 
When you import a CSV file or an Excel sheet into Python, the computer essentially sees it as a giant version of this "List of Dictionaries."

Key Advantages
Scalability: 
You can easily add a new customer by just adding one more dictionary to the list: customers.append({"name": "Leo", "revenue": 300}).

Searchability: 
You can use a for loop to go through the list and find only customers where revenue > 100.

Organization: 
It keeps related data (like a name and their specific revenue) tied together so they don't get mixed up with other people's data.

### Removing a Column

```python
df.drop("city", axis=1)
```

axis=1: This is crucial. In Pandas, axis=0 refers to rows, and axis=1 refers to columns. 

By saying axis=1, you are telling the computer, "Find the vertical column named 'city' and remove it."

```python
df = df.drop("city", axis=1)
```

```python
df
```

df.drop("city", axis=1)  VS    df = df.drop("city", axis=1)

When you just run df.drop("city", axis=1), Pandas shows you a preview of what the table looks like without that column.

However, it doesn't actually change the original df variable.

If you typed df again right after, the "city" column would still be there!


To actually save your changes, you have to use Reassignment: df = df.drop("city", axis=1)

This tells the computer: "Take the version of the table without the city column and save it over the old version of df."
