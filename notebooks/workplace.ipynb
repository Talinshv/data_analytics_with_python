```python
revenues = [120, 250, 80, 310, 95]
```

```python
# Create a new list with revenues after applying 10% tax.
new_list =[value*0.9 for value in revenues]
new_list
```

```python
# Create a list containing only revenues greater than 100.
new_list = [value for value in revenues if value >100]
new_list
```

```python
# Create a list labeling each revenue as "High" if > 200, otherwise "Normal".
new_list = ["high" if value > 200 else "normal" for value in revenues]
new_list
```

```python
import pandas as pd
SQL_URL = "https://raw.githubusercontent.com/Talinshv/sql-analytics-portfolio/refs/heads/main/data/public_schema/"
TO_SAVE_URL = 'C:/ACA/sql-analytics-portfolio/data/public_schema'
l = ['customers', 'employees','orders','products', 'sales']

for i in l:
    print(100*'-')
    print(f"trying to download {i}.csv")
    df = pd.read_csv(SQL_URL+i+'.csv')
    print(f"{i}.csv is downloaded")
    print(f"the shape of {i}.csv : {df.shape}")
    print(f"trying to save  {i}.csv")
    df.to_csv(TO_SAVE_URL+i+'.csv',index=False)
    print(f"{i}.csv is successfully saved")
```
